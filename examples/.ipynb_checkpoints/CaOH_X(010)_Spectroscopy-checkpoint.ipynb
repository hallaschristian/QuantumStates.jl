```julia
using QuantumStates
using LinearAlgebra
using Parameters
using DataFrames
using PrettyTables
```

```julia
const c = 299792458
const h = 6.62607015e-34
const ħ = h / 2π
;
```

```julia
# Spectroscopic constants for X(010)
# From Table III, Li and Coxon (1995)
# Units are MHz
@consts begin
    BX = 0.333458 * c * 1e-4
    DX = 0.3930e-6 * c * 1e-4
    γX = 0.001184 * c * 1e-4
    qX = -0.7181e-3 * c * 1e-4
    bFX = 2.602
    cX = 2.053
#     bFX = 2.45 # 2.602
#     cX = 2.6 # 2.053
    μX = 1e-6 * 1.465 * 10^2 * (1e-21 / c) / h # 1e-6 = Hz to MHz
end
;
```

```julia
# Define the zero-field Hamiltonian operator
H_bare(state, state′, i, j, p) = 
    (BX * Rotation + 
    DX * RotationDistortion + 
    γX * SpinRotation + 
    bFX * Hyperfine_IS + 
    (cX/3) * Hyperfine_Dipolar +
    qX * ℓDoubling)(state, state′)
;
```

```julia
QN_bounds = (S = 1/2, I = 1/2, Λ = [-1,1], N = 1:3)
basis = enumerate_states(HundsCaseB_Rot, QN_bounds)
ordered_basis = QuantumStates.order(basis, :M)
;
```

```julia
p = NamedTuple()
H = Hamiltonian(H_operator=H_bare, basis=ordered_basis)
update_matrix!(H, p)
solve!(H)
;
```

## Add an E-field

```julia
const gS = 2.0023193043617
const μB = 1.39962
;
```

```julia
Stark_M = zeros(length(H.basis), length(H.basis))
for (i, state) ∈ enumerate(H.basis)
    for (j, state′) ∈ enumerate(H.basis)
        Stark_M[i,j] = Stark(state, state′)
    end
end
Zeeman_M = zeros(length(H.basis), length(H.basis))
for (i, state) ∈ enumerate(H.basis)
    for (j, state′) ∈ enumerate(H.basis)
        Zeeman_M[i,j] = Zeeman(state, state′)
    end
end

p = (M0 = H.M, Stark_M = Stark_M, Zeeman_M = Zeeman_M)

H_Efield_op(state, state′, i, j, p, E, B) = 
    p.M0[i,j] + μX * E * p.Stark_M[i,j] + gS * μB * B * p.Zeeman_M[i,j]
;
```

```julia
H_Efield = Hamiltonian(H_operator=H_Efield_op, basis=ordered_basis)
E = 200 * 0.3
B = 0.0
update_matrix!(H_Efield, p, E, B)
solve!(H_Efield)
;
```

```julia
transitions = compute_transitions(H_Efield, 0)
transitions = [transitions; compute_transitions(H_Efield, 1)]
;
```

```julia
df_transitions = transitions_table(transitions)
df_transitions_clean = select(df_transitions, 
    Not([:State, :State_1, :Λ, :I, :S, :E, :c_1, :Λ_1, :I_1, :S_1, :E_1]))
filter!(:f => f -> 39800 < f < 40100, df_transitions_clean)
# filter!(:J => J -> J == 3//2, df_transitions_clean)
filter!(:F => F -> F == 0, df_transitions_clean)
# filter!(:F_1 => F -> F == 1, df_transitions_clean)
# filter!(:J_1 => J -> J == 3//2, df_transitions_clean)
# filter!(:M_1 => M -> M == 0, df_transitions_clean)
df_transitions_clean.f .-= 39960
# df_transitions_clean.f ./= 2
df_transitions_clean.f .+= 1.5
pretty_table(df_transitions_clean, nosubheader=true)
```

## Add polarizability

```julia
α = [-5.84517*10^-39, 0, -1.2469*10^-39];
```

```julia
# Trap characteristics
w = 25 * 10e-6
P = 13.25 * (1.5 / 3.5)
I_trap = 2P / (π * w^2)
;
```

```julia
Stark_M = zeros(length(H.basis), length(H.basis))
for (i, state) ∈ enumerate(H.basis)
    for (j, state′) ∈ enumerate(H.basis)
        Stark_M[i,j] = Stark(state, state′)
    end
end
Zeeman_M = zeros(length(H.basis), length(H.basis))
for (i, state) ∈ enumerate(H.basis)
    for (j, state′) ∈ enumerate(H.basis)
        Zeeman_M[i,j] = Zeeman(state, state′)
    end
end
polarizability_M = zeros(length(H.basis), length(H.basis))
for (i, state) ∈ enumerate(H.basis)
    for (j, state′) ∈ enumerate(H.basis)
        polarizability_M[i,j] = (10e-6 / h) * polarizability(state, state′, α,  0)
#         polarizability_M[i,j] += (10e-6 / h) * polarizability(state, state′, α, +1)
#         polarizability_M[i,j] += (10e-6 / h) * polarizability(state, state′, α, -1)
    end
end

p = (M0 = H.M, Stark_M = Stark_M, Zeeman_M = Zeeman_M, polarizability_M = polarizability_M)

H_full_op(state, state′, i, j, p, E, B, I) = 
    p.M0[i,j] + μX * E * p.Stark_M[i,j] + gS * μB * B * p.Zeeman_M[i,j] + I * p.polarizability_M[i,j]
;
```

```julia
H_full = Hamiltonian(H_operator=H_full_op, basis=ordered_basis)
E = 61.0
B = 0.300
I = I_trap
update_matrix!(H_full, p, E, B, I)
solve!(H_full)

transitions = compute_transitions(H_full, 0)
transitions = [transitions; compute_transitions(H_full, -1)]
transitions = [transitions; compute_transitions(H_full, 1)]
;
```

### EDM state splitting between M=-1, M=+1
