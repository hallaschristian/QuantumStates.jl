##### Load $\widetilde{X}(01^10)$ Hamiltonian

```julia
@time H0 = load_from_file("CaOH_BendingMode_Hamiltonian", "X://My Drive//github//QuantumStates//Hamiltonians//")
;
```

```julia
import QuantumStates: polarizability, Zeeman

# const α = [1.6e-40, 0, 2.7e-38] / (2ε0 * c) / h # 573 nm, in units of h * Hz * W^(-1) * m^(-2) * (2ε0 * c) before conversion; Hz * W^(-1) m^(-2) after
const α = [3.37471e-39, 0, -1.01809e-39] / (2ε0 * c) / h # 1064 nm
const θ = 0 * (π / 2)
const ϵ_trap = cos(θ) * [0.0, 1.0, 0.0] + sin(θ) * [1.0, 0.0, -1.0] / √2
polarizability(state, state′) = polarizability(state, state′, α, ϵ_trap)
@time H = add_to_H(H0, :I_trap, 1e-6 * 1e9 * polarizability) # 1e-6 to convert from Hz to MHz, 1e9 from W to GW

@time H = add_to_H(H, :E, (1e2 * μX) * Stark) # 1e2 converts from V/m to V/cm

const B_x = [1.0, 0.0, -1.0] ./ √2
const B_z = [0.0, 1.0, 0.0]
Zeeman_x(state, state′) = Zeeman(state, state′, B_x)
Zeeman_z(state, state′) = Zeeman(state, state′, B_z)
@time H = add_to_H(H, :B_x, gS * μB * Zeeman_x)
@time H = add_to_H(H, :B_z, gS * μB * Zeeman_z)

H.parameters.I_trap = 0.0
H.parameters.E = 500.0 # 198 * 0.3
H.parameters.B_x = 0.0
H.parameters.B_z = 0.001

@time full_evaluate!(H)
@time solve!(H)
;
```

### Tensor Stark shifts

```julia
I_trap_scan_range = (0:0.1:20)
x = scan_single_parameter(deepcopy(H), :I_trap, I_trap_scan_range);
```

```julia
trap_shifts = zeros(length(I_trap_scan_range), length(H.states))
for (i, I_trap_value) ∈ enumerate(sort(collect(keys(x.state_dict))))
    trap_shifts[i,:] = energy.(x.state_dict[I_trap_value])
end
```

```julia
N2_idxs = subspace(H.states, (N=1,))[1];
```

```julia
plot(I_trap_scan_range, trap_shifts[:, N1_idxs], legend=nothing)
```

It appears that states with $M_N = 0$ have positive tensor shifts, while those with $|M_N| = 1$ have negative tensor shifts.

### How deep do the tweezers need to be?

Let's assume we want a tweezer with a trap depth of $\sim 100~\mu\text{K}$. For a scalar trap at $1064~\text{nm}$, this will require an intensity of $\sim 2~\text{GW m}^{-2}$. Roughly the same order of magnitude intensity is required of the tensor trap, based on trap shifts, though it'll need to be significantly deeper than the scalar trap to ensure that the tensor-trapped molecule does not tunnel to the scalar trap.

```julia
ω_r = 233e3
T = 10e-6
k_B = 1.38e-23
amu = 1.67e-27
m = 57 * amu
σ_r = (1 / ω_r) * sqrt(k_B * T / m)
σ_r * 1e6
```

```julia
k2 = 1e-10 * 1e-6
r0 = 60e-9
z0 = 0.2e-6
r = 200e-9
k2 * exp(-r^2 / (2 * r0^2)) * (1 / ((2π)^3 * r0^2 * z0))
```

### Dipole moments at given trap depths

```julia
function dipole_moment(H)
    original_E = H.parameters.E

    ΔE = 1e-5

    H.parameters.E = original_E + (- ΔE)
    evaluate!(H)
    solve!(H)
    energies_m = energy.(H.states)

    H.parameters.E = original_E + (+ ΔE)
    evaluate!(H)
    solve!(H)
    energies_p = energy.(H.states)

    dipole_moments = (energies_p .- energies_m) ./ ΔE

    return dipole_moments
end; 
```

```julia
y = scan_single_parameter(deepcopy(H), :I_trap, I_trap_scan_range, dipole_moment)
;
```

```julia
plot(I_trap_scan_range, y[:,17:24], legend=nothing, size=(1200,700))
```
