## Compute histogram of the trap depths explored by the molecules

What is the range of trap depths explored by the molecules? The density is $\propto \exp(-x^2/2\sigma_x^2) \exp(-y^2/2\sigma_y^2) \exp(-z^2/2\sigma_z^2)$, where $\sigma_\alpha = \omega_\alpha^{-1} \sqrt{k_B T/m}$, and we'll use a temperature of $T = 60~\mu\text{K}$.

```julia
amu = 1.66e-27
kB = 1.38e-23
m = 57 * amu
ωx = 2π * 6774/2
ωy = 2π * 8721/2
ωz = 218.734
temp = 60e-6
const w0x = 0.0000275
const w0y = 0.0000233
const λ = 1064e-9
const zRx = π * w0x^2 / λ
const zRy = π * w0y^2 / λ
σ_x = ωx^(-1) * sqrt(kB * temp / m)
σ_y = ωy^(-1) * sqrt(kB * temp / m)
σ_z = ωz^(-1) * sqrt(kB * temp / m)
;
```

```julia
using Distributions

waist(z,w0,zR,z0) = w0 * sqrt(1 + ((z-z0) / zR)^2)
const T_depth = (1.0 / 3.5) * 645e-6
function ODT_Potential(r)
    x, y, z = r
    wx = waist(z, w0x, zRx, 0)
    wy = waist(z, w0y, zRy, 0)    
    return T_depth - T_depth * (1 - ((w0x * w0y) / (wx * wy)) * exp(-2 * (x^2 / wx^2 + y^2 / wy^2)))
end
r = (Normal(0, σ_x), Normal(0, σ_y), Normal(0, σ_z))

n = 10_000_000
rs = Vector{Float64}[]
@time for i ∈ 1:n
    push!(rs, [rand(r_i) for r_i ∈ r])
end

depths = ODT_Potential.(rs)

mean(depths) .* 1e6 |> display
std(depths) .* 1e6 |> display
```

```julia
using StatsBase, LsqFit
```

```julia
hist = fit(Histogram, depths ./ T_depth, 0:0.001:1.0) 
plot(hist, legend=nothing)
```

```julia
# Let's make our own implementation for linear interpolation in a regular grid with an arbitrary number of dimensions
struct LinearInterpolation{N}
    scales::NTuple{N, Float64}
    offsets::NTuple{N, Float64}
    grid::Array{Float64, N}
    dims::NTuple{N, Int64}
end
function LinearInterpolation(knots, grid)
    scales = zeros(length(knots))
    offsets = zeros(length(knots))
    for i ∈ eachindex(knots)
        scales[i] = 1 / step(knots[i])
        offsets[i] = minimum(knots[i]) - step(knots[i])
    end
    LinearInterpolation(tuple(scales...), tuple(offsets...), grid, size(grid))
end
function (itp::LinearInterpolation)(x)::Float64
    # print(x)
    grid_point = (x .- itp.offsets) .* itp.scales
    grid_point_idxs = floor.(Int64, grid_point)
    idxs = Iterators.product(((grid_point_idxs[i], grid_point_idxs[i] + 1) for i ∈ eachindex(grid_point_idxs))...)
    val = 0.0
    for _x ∈ idxs
        if all(_x .<= itp.dims)
            val += itp.grid[_x...] * reduce(*, 1 .- abs.(grid_point .- _x))
        end
    end    
    return val
end
lin_itp = LinearInterpolation(knots, grid)
;
```

```julia
const AC_noise = 3.0e-3
const AC_phase = 0.0 #π/3
function cos_integrate(x, p)
    Bx, By, Bz, I_trap = x
    itp, itp_I_trap, E, Bx_μ, By_μ, Bz_μ, I_trap_μ, t = p
    ω = 2π * itp(E, Bx, By, Bz, I_trap)
    return (
        itp_I_trap(I_trap) * Bx_weight(Bx - Bx_μ) * By_weight(By - By_μ) * Bz_weight(Bz - Bz_μ) 
        * (1 + cos(
                ω * t
                + (AC_noise / Bz_μ) * ω * sin(AC_phase + 2π * 60 * t) / (2π * 60) # AC noise term
                )
            ) / 2
        )
end
;
```

```julia
# Contruct the `pd` for the B field noise
const Bx_σ = 0.1e-3
const By_σ = 0.1e-3
const Bz_σ = 0.1e-3
const normal_dist_x = Normal(0.0, Bx_σ)
const normal_dist_y = Normal(0.0, By_σ)
const normal_dist_z = Normal(0.0, Bz_σ)
Bx_weight(x) = pdf(normal_dist_x, x)
By_weight(x) = pdf(normal_dist_y, x)
Bz_weight(x) = pdf(normal_dist_z, x)
;
```

```julia
# Parameters
E = 201 * 0.3
Bx = +3.0e-3
By = +3.0e-3
Bz = +2.5e-3
I_trap = 0.6
t = 50.0

p = (itp, itp_I_trap, E, Bx, By, Bz, I_trap, t)
a = (Bx - 3Bx_σ, By - 3By_σ, Bz - 3Bz_σ, 0.5)
b = (Bx + 3Bx_σ, By + 3By_σ, Bz + 3Bz_σ, 1.0)
prob = IntegralProblem{false}(cos_integrate, a, b, p)
;
```

```julia
using ProgressMeter, IntegralsCuba
times = (0:0.2:100) .* 1e-3
vals = zeros(length(times))

progress = Progress(length(times))
@time Threads.@threads for i ∈ eachindex(times)
    t = times[i]
    a = (Bx - 3Bx_σ, By - 3By_σ, Bz - 3Bz_σ, 0.5)
    b = (Bx + 3Bx_σ, By + 3By_σ, Bz + 3Bz_σ, 1.0)
    prob_t = remake(prob, lb=a, ub=b, p=(itp, itp_I_trap, E, Bx, By, Bz, I_trap, t))
    vals[i] = solve(prob_t, CubaDivonne(), abstol=1e-3).u
    next!(progress)
end
```

```julia
using Plots
@. decaying_cosine_gaussian(t, p) = (1 + cos(2π * t / p[1]) * exp(-t^2 / p[2]^2)) / 2
@. decaying_cosine(t, p) = (1 + cos(2π * t / p[1]) * exp(-t / p[2])) / 2
fit_func = decaying_cosine

p0 = [25e-3, 10e-3]
fitted_data = curve_fit(fit_func, times, vals, p0)
display(fitted_data.param)
plot(times, vals, legend=nothing, ylim=(0,1))
plot!(times, fit_func(times, fitted_data.param))
```
