### Zeeman plot

```python
using LaTeXStrings, Plots
```

```python
function H_func!(H, scan_values)
    H.parameters.B_z = scan_values[1]
    evaluate!(H)
    QuantumStates.solve!(H)
    return nothing
end
;
```

```python
Bs = 0:100.:10000
scan_params = (
    B_z = Bs,
    );
iterator = Iterators.product
@time scan_values, _ = scan_parameters(deepcopy(H), scan_params, iterator, H_func!, H -> energy.(H.states), n_threads=1)
matrix_values = hcat(values(scan_values)...)';
;
```

```python
plot(Bs, matrix_values[:,5:40], linewidth=1.5, linecolor=:blue, alpha=0.5)
plot!(
    legend=nothing, 
    # size=(1400,700),
    xlabel="Magnetic field (G)",
    ylabel="Energy (MHz)",
    labelfontsize=13,
    tickfontsize=13,
    grid=:off,
    box=:on
    )
```
