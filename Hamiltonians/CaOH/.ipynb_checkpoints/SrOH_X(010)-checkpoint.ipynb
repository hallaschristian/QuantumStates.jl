```julia
using QuantumStates, UnitsToValue
```

```julia
H_operator = :(
    BX * Rotation + 
    DX * RotationDistortion + 
    γX * SpinRotation_Λ0 +
    qX * ℓDoubling
);
```

```julia
# From (1993)
parameters = @params begin
    BX = 0.2485815 * 2.99792458e8 * 1e-4
    DX = 2.2097e-7 * 2.99792458e8 * 1e-4
    γX = 2.409e-3 * 2.99792458e8 * 1e-4
    qX = -3.9503e-4 * 2.99792458e8 * 1e-4
end;
```

```julia
QN_bounds = (S = 1/2, I = 1/2, Λ = [-1,1], N = 1:3)
basis = enumerate_states(HundsCaseB_Rot, QN_bounds)
;
```

```julia
SrOH_X010 = @time Hamiltonian(basis=basis, operator=H_operator, parameters=parameters)
@time full_evaluate!(SrOH_X010)
@time solve!(SrOH_X010)
;
```

```julia
@time save_to_file(SrOH_X010, "SrOH_X010", "")
```
